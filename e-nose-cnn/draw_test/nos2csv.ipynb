```python
import os
import pandas as pd
import numpy as np
import random


def listdir(path, list_name):  # 传入存储的list
    for file in os.listdir(path):
        file_path = os.path.join(path, file)
        if os.path.isdir(file_path):
            listdir(file_path, list_name)
        else:
            list_name.append(file_path)


file_paths = []
listdir("F:\\gitworkspace\\python\\e-nose-cnn\\draw_test", file_paths)
#print(file_paths)

for i in file_paths:
    # excel_data = pd.read_csv(i,sep=',',dtype=np.object,userows=[44, 104])
    # excel_data = pd.read_csv(i,sep=' ',dtype=np.object,skiprows=52,nrows=200,header=None)
    try:
        data = pd.read_csv(i, sep='\t', dtype=np.object,
            skiprows=52, header=None, engine='python')
        data = data.loc[0:119]
        # data = data[range(10)]
        data = data[random.sample(range(0,10),10)]
        # print(data)
        data.to_csv(os.path.join("F:\\gitworkspace\\python\\e-nose-cnn\\draw_test", "{0[4]}-1.csv".format(
            i.split("\\"))), encoding='utf_8_sig', index=False, header=1)  # 单纯utf8编码用excel打开会乱码
    except Exception as e:
        print(i)
```
