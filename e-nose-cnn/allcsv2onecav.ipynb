```python
import os
import pandas as pd
import numpy as np


def listdir(path, list_name):  # 传入存储的list
    for file in os.listdir(path):
        file_path = os.path.join(path, file)
        if os.path.isdir(file_path):
            listdir(file_path, list_name)
        else:
            list_name.append(file_path)

file_paths = []
listdir("F:\\gitworkspace\\python\\e-nose-cnn\\train\\all_csv\\4", file_paths)
# print(file_paths)

for i in file_paths:
    try:
        df = pd.read_csv(i)
        data = df.values
        data = list(map(list,zip(*data)))
        data = pd.DataFrame(data)
        data.to_csv(os.path.join("F:\\gitworkspace\\python\\e-nose-cnn\\train", "{0[4]}.csv".format(
            i.split("\\"))), encoding='utf_8_sig',index=False, header=0, mode = 'a+')  # 单纯utf8编码用excel打开会乱码
    except Exception as e:
        print(i)
```
