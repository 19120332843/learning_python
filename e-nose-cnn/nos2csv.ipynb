```python
import os
import pandas as pd
import numpy as np


def listdir(path, list_name):  # 传入存储的list
    for file in os.listdir(path):
        file_path = os.path.join(path, file)
        if os.path.isdir(file_path):
            listdir(file_path, list_name)
        else:
            list_name.append(file_path)

file_paths = []
listdir("F:\\gitworkspace\\python\\e-nose-cnn\\nos-data\\binglang2\\2", file_paths)
# print(file_paths)

for i in file_paths:
    try:
        #nos2csv
        data = pd.read_csv(i, sep='\t', dtype=np.object,
            skiprows=52, header=None, engine='python')
        # print(data)
        data = data.loc[0:119]
        # print(data)
        data = data[range(10)]
        # print(data)
        data.to_csv(os.path.join("F:\\gitworkspace\\python\\e-nose-cnn\\test\\all_csv2\\1", "{0[7]}-test.csv".format(
            i.split("\\"))), encoding='utf_8_sig', index=False, header=0)#utf_8_sig
    except Exception as e:
        print(i)
```
