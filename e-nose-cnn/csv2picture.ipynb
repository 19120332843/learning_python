```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def Normlize(Z):
    Zmax, Zmin = Z.max(axis=1), Z.min(axis=1)
    Zmean = Z.mean(axis=1)
    #按列排序
    Zmax, Zmin = Zmax.reshape(-1, 1), Zmin.reshape(-1, 1)
    Zmean = Zmean.reshape(-1, 1)
    Z = (Z - Zmean) / (Zmax - Zmin)
    return Z

pic = pd.read_csv("F:\\gitworkspace\\python\\e-nose-cnn\\draw_test\\draw_test.csv")
# print(pic)

print(pic)

np.save('F:\\gitworkspace\\python\\e-nose-cnn\\draw_test\\draw_test.npy', pic)

pic = np.load('F:\\gitworkspace\\python\\e-nose-cnn\\draw_test\\draw_test.npy')

pic = Normlize(pic)

print(pic)

np.savetxt('F:\\gitworkspace\\python\\e-nose-cnn\\draw_test\\new.xlsx', pic, delimiter = ',') 

pic = pd.read_excel("F:\\gitworkspace\\python\\e-nose-cnn\\draw_test\\new.xlsx")

data = range(10)
data = pd.DataFrame(data)
data = data.values
data = list(map(list,zip(*data)))
data = pd.DataFrame(data)
data.to_csv(os.path.join("F:\\gitworkspace\\python\\e-nose-cnn\\draw_test", "new1.csv"), encoding='utf_8_sig', index=False, header=0)

pic.to_csv(os.path.join("F:\\gitworkspace\\python\\e-nose-cnn\\draw_test", "new1.csv"), encoding='utf_8_sig',index=False, header=0, mode = 'a+') 

pic = pd.read_csv("F:\\gitworkspace\\python\\e-nose-cnn\\draw_test\\new1.csv")

print(pic)

x = np.arange(0, 120, 1)

y0 = pic['0']
y1 = pic['1']
y2 = pic['2']
y3 = pic['3']
y4 = pic['4']
y5 = pic['5']
y6 = pic['6']
y7 = pic['7']
y8 = pic['8']
y9 = pic['9']

plt.xlim(-1, 130)
plt.ylim(0, 11)

plt.plot(x, y0)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.plot(x, y5)
plt.plot(x, y6)
plt.plot(x, y7)
plt.plot(x, y8)
plt.plot(x, y9)

plt.show()
```
