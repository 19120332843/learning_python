```python
def Normlize(Z):
    Zmax, Zmin = Z.max(axis=1), Z.min(axis=1)
    print(Zmax)
    print(Zmin)
    Zmean = Z.mean(axis=1)
    print(Zmean)
    Zmax, Zmin = Zmax.reshape(-1, 1), Zmin.reshape(-1, 1)
    print(Zmax)
    print(Zmin)
    Zmean = Zmean.reshape(-1, 1)
    Z = (Z - Zmean) / (Zmax - Zmin)
    return Z

train_x_Normed = Normlize(train_x)

train_x_Normed
```

```python
train_x = torch.from_numpy(train_x_Normed).type(torch.cuda.FloatTensor)

train_x
```
